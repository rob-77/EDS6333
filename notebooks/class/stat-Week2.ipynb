```python
import matplotlib.pyplot as plt

age1 = [28, 44, 22, 31, 34, 48, 45, 32, 24, 32,
        36, 46, 46, 26, 37, 32, 62, 31, 30, 35]

plt.hist(age1, bins=5, edgecolor="black", alpha=0.7)

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Ages")

plt.show()
```
