```python
from scipy.stats import binom

n = 5
p = 0.2
x = 2  # exactly 2 crashes

pmf_value = binom.pmf(x, n, p)
cdf_value = binom.cdf(x, n, p)

print(f"Exactly 2 => PMF (x={x}) = {pmf_value:.4f}")
print(f"At most 2 => CDF (x<={x}) = {cdf_value:.4f}")


x= 1 # at most 1 crash
cdf_value = binom.cdf(x, n, p)  
print(f"At most 1 => CDF (x<={x}) = {cdf_value:.4f}")        

```

Question: A batch of 25 microchips (N=25) contains 5 defective chips (R=5). If a sample of 10 chips (n=10) is selected without replacement, what is the probability that exactly 2 are defective (x=2)?

```python
from scipy.stats import hypergeom

N = 25   # population size
R = 5    # number of defective chips
n = 10    # sample size
x = 2    # observed defective chips

pmf_value = hypergeom.pmf(x, N, R, n)
print(f"PMF (x={x}) = {pmf_value:.4f}")


```

A server receives an average of 4 error pings per minute (lamda=4). What is the probability that the server receives exactly 6 error pings in a given minute?

```python
from scipy.stats import poisson

mu = 4   # average rate
x = 6    # observed events

pmf_value = poisson.pmf(x, mu)
print(f"PMF (x={x}) = {pmf_value:.4f}")
```

Question: A sales rep has a 10% chance of making a sale on any given cold call (p=0.1). What is the probability that they make their first sale exactly on the 5th call?

```python
from scipy.stats import geom

p = 0.1
x = 5

pmf_value = geom.pmf(x, p)
print(f"PMF (x={x}) = {pmf_value:.4f}")
```

A medicine is effective 70% of the time (p=0.7). What is the probability that 9 volunteers (x=9) are needed to achieve the 7th recovery (r=7)?

```python
from scipy.stats import nbinom

x = 9     #number of volunteers
r = 7     # number of successes
p = 0.7   # success probability
# k = x - r
k = x- r     # number of failures before r-th success

pmf_value = nbinom.pmf(k, r, p)
print(f"PMF (k={k}) = {pmf_value:.4f}")
```
