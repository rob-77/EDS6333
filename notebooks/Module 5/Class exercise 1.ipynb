### A professor assigns a random presentation slot to 10 students by drawing a number from 1 to 10 from a hat. What is the probability that the assigned slot is greater than 7? 

```python
favorable = [8, 9, 10]
total = 10

prob = len(favorable) / total
print(f"{prob:.4f}")
```

```python
from scipy.stats import randint

prob = randint.cdf(10, 1, 11) - randint.cdf(7, 1, 11)
print(f"{prob:.4f}")
```
