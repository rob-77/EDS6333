A quality inspector checks a single motherboard from a production line. The historical defect rate is 5%. Let "finding a defect" be a Success (p=0.05). What is the probability the board is defective, and what is the variance of this distribution?

```python
p=0.05
mean     = p
variance = p * (1 - p)

print(f"mean     = p           = {mean}")
print(f"variance = p * (1 - p) = {p} × {1-p} = {variance}")
```
