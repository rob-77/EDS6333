## Binomial
You can compute both the PMF and CDF of a binomial distribution in SciPy by calling  and . The PMF gives the probability of exactly  successes, while the CDF gives the probability of up to and including  successes.

This example uses n=10, p=0.4, and evaluates at x=3.

```python
from scipy.stats import binom

n = 10
p = 0.4
x = 3

pmf_value = binom.pmf(x, n, p)
cdf_value = binom.cdf(x, n, p)

print(f"PMF (x={x}) = {pmf_value:.4f}")
print(f"CDF (x<={x}) = {cdf_value:.4f}")
```

What these values mean
- PMF tells you the probability of getting exactly 3 successes in 10 trials.
- CDF tells you the probability of getting 0, 1, 2, or 3 successes in 10 trials.

## Hypergeometric
The hypergeometric distribution models sampling without replacement, making it ideal for teaching scenarios like drawing cards, selecting defective items, or choosing colored balls from an urn. The SciPy functions hypergeom.pmf and hypergeom.cdf let you compute exact probabilities for these situations.
🎯 Core idea with parameters
- N — total population size
- R — number of “success” items in the population
- n — number of draws
- x — number of observed successes

The PMF gives the probability of getting exactly x successes; the CDF gives the probability of getting up to and including x successes.



Note the parameter names differ, but the meaning is the same: 𝑁=population, 𝑅=successes in
population, 𝑛=draws, 𝑥=successes in sample

Scenario:
A box has N = 20 items, R = 7 of them are defective.
You randomly inspect n = 5 items.
What is the probability that x = 2 of them are defective?

```python
from scipy.stats import hypergeom

N = 20   # population size
R = 7    # number of success states in population
n = 5    # number of draws
x = 2    # observed successes

pmf_value = hypergeom.pmf(x, N, R, n)
cdf_value = hypergeom.cdf(x, N, R, n)

print(f"PMF (x={x}) = {pmf_value:.4f}")
print(f"CDF (x<={x}) = {cdf_value:.4f}")
```

 Interpretation
- 	PMF = 0.3241 → There is about a 32.41% chance of finding exactly 2 defective items in your 5‑item sample.
- 	CDF = 0.7905 → There is about a 79.05% chance of finding 0, 1, or 2 defective items

Full Distribution Table

```python
import numpy as np

xs = np.arange(0, n+1)
pmf_all = hypergeom.pmf(xs, N, R, n)
cdf_all = hypergeom.cdf(xs, N, R, n)

for xi, pm, cd in zip(xs, pmf_all, cdf_all):
    print(f"x={xi:2d}  PMF={pm:.4f}  CDF={cd:.4f}")
```

## Poisson
A Poisson example in SciPy centers on a single rate parameter \mu =\lambda , which represents the average number of events in a fixed interval. The PMF gives the probability of exactly x events, and the CDF gives the probability of up to and including x events. The SciPy documentation confirms the available methods, including pmf and cdf, and the standard PMF formula f(k)=e^{-\mu }\mu ^k/k!.

Parameters
- 	x — number of observed events
- 	mu (λ) — average event rate (mean and variance of the distribution)

Scenario:
A call center receives an average of λ = 4 calls per hour.
What is the probability of receiving exactly 2 calls, and what is the probability of receiving 2 or fewer calls?

```python
from scipy.stats import poisson

mu = 4   # average rate
x = 2    # observed events

pmf_value = poisson.pmf(x, mu)
cdf_value = poisson.cdf(x, mu)

print(f"PMF (x={x}) = {pmf_value:.4f}")
print(f"CDF (x<={x}) = {cdf_value:.4f}")
```

Interpretation
- 	PMF = 0.1465 → About a 14.65% chance of receiving exactly 2 calls in an hour.
- 	CDF = 0.2381 → About a 23.81% chance of receiving 0, 1, or 2 calls.

Full distribution table

```python
import numpy as np

xs = np.arange(0, 10)
pmf_all = poisson.pmf(xs, mu)
cdf_all = poisson.cdf(xs, mu)

for xi, pm, cd in zip(xs, pmf_all, cdf_all):
    print(f"x={xi:2d}  PMF={pm:.4f}  CDF={cd:.4f}")
```

## Geometric

If your X = trial number of the 1st success (1,2,3,…):

A geometric example in SciPy uses the idea that the distribution models the trial number on which the first success occurs.

Scenario:
A process has a success probability p = 0.3 on each independent trial.
What is the probability that the first success occurs on trial x = 4, and what is the probability that it occurs on or before trial 4?

```python
from scipy.stats import geom

p = 0.3
x = 4

pmf_value = geom.pmf(x, p)
cdf_value = geom.cdf(x, p)

print(f"PMF (x={x}) = {pmf_value:.4f}")
print(f"CDF (x<={x}) = {cdf_value:.4f}")
```

 Interpretation
- 	PMF gives the probability that the first success happens exactly on trial 4.
- 	CDF gives the probability that the first success happens on trial 1, 2, 3, or 4.


Full distribution table

```python
import numpy as np

xs = np.arange(1, 8)
pmf_all = geom.pmf(xs, p)
cdf_all = geom.cdf(xs, p)

for xi, pm, cd in zip(xs, pmf_all, cdf_all):
    print(f"x={xi:2d}  PMF={pm:.4f}  CDF={cd:.4f}")
```

## Negative Binomial (r-th success)
Most software uses k = failures before r-th success, so if your problem gives trial number 𝑥 then use: k = x - r.

The negative binomial distribution in SciPy models the number of failures  before achieving r successes, with each trial having success probability . The PMF and CDF let you compute the probability of getting exactly  failures or up to  failures before the -th success.

📘 Parameters and meaning
- r — number of successes you are waiting for
- p — probability of success on each trial
- k — number of failures before the r‑th success

Scenario:
A process has a success probability p = 0.3 on each independent trial.
What is the probability that the first success occurs on trial x = 4, and what is the probability that it occurs on or before trial 4?

```python
from scipy.stats import nbinom

r = 2     # number of successes
p = 0.4   # success probability
k = 3     # number of failures before r-th success

pmf_value = nbinom.pmf(k, r, p)
cdf_value = nbinom.cdf(k, r, p)

print(f"PMF (k={k}) = {pmf_value:.4f}")
print(f"CDF (k<={k}) = {cdf_value:.4f}")
```

Interpretation
- PMF = 0.1382 → About a 13.82% chance of seeing exactly 3 failures before the 2nd success.
- CDF = 0.6826 → About a 68.26% chance of seeing 0, 1, 2, or 3 failures before the 2nd success.

Full distribution table

```python
import numpy as np

ks = np.arange(0, 8)
pmf_all = nbinom.pmf(ks, r, p)
cdf_all = nbinom.cdf(ks, r, p)

for ki, pm, cd in zip(ks, pmf_all, cdf_all):
    print(f"k={ki:2d}  PMF={pm:.4f}  CDF={cd:.4f}")
```
