## Binomial

### binom.pmf(x,n,p) (PMF), binom.cdf(x,n,p) (CDF)

```python
from scipy.stats import binom

# This example uses n=10, p=0.4, and evaluates at x=3

n = 10
p = 0.4
x = 3

pmf_value = binom.pmf(x, n, p)
cdf_value = binom.cdf(x, n, p)

print(f"PMF (x={x}) = {pmf_value:.4f}")
print(f"CDF (x<={x}) = {cdf_value:.4f}")
```

## Bernoulli

```python
from scipy.stats import binom

p = 0.3   # probability of success
n = 1     # one trial

# x = 0 (failure)
pmf_0 = binom.pmf(0, n, p)
print(f"pmf(x=0) = {pmf_0:.4f}")

# x = 1 (success)
pmf_1 = binom.pmf(1, n, p)
print(f"pmf(x=1) = {pmf_1:.4f}")
```

## Hypergeometric

Scenario:
A box has N = 20 items, R = 7 of them are defective.
You randomly inspect n = 5 items.
What is the probability that x = 2 of them are defective?

```python
from scipy.stats import hypergeom

N = 20   # population size
R = 7    # number of success states in population
n = 5    # number of draws
x = 2    # observed successes

pmf_value = hypergeom.pmf(x, N, R, n)
cdf_value = hypergeom.cdf(x, N, R, n)

print(f"PMF (x={x}) = {pmf_value:.4f}")
print(f"CDF (x<={x}) = {cdf_value:.4f}")
```
